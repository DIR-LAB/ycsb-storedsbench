```python
text_file = open("../results/skiplist-benchmark.out", "r")

lines = text_file.readlines()
print("# of lines: ", len(lines))
text_file.close()
```

```python
line_it = 0

#dram workload a
dram_wa = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        dram_wa.append(val)
print(dram_wa)

#dram workload b
dram_wb = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        dram_wb.append(val)
print(dram_wb)

#dram workload c
dram_wc = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        dram_wc.append(val)
print(dram_wc)

#dram workload d
dram_wd = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        dram_wd.append(val)
print(dram_wd)

#dram workload e
dram_we = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        dram_we.append(val)
print(dram_we)

######################[dram data loaded]######################

#pmem workload a
pmem_wa = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_wa.append(val)
print(pmem_wa)

#pmem workload b
pmem_wb = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_wb.append(val)
print(pmem_wb)

#pmem workload c
pmem_wc = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_wc.append(val)
print(pmem_wc)

#pmem workload d
pmem_wd = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_wd.append(val)
print(pmem_wd)

#pmem workload e
pmem_we = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_we.append(val)
print(pmem_we)

######################[pmem data loaded]######################

#pmem_tx workload a
pmem_tx_wa = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_tx_wa.append(val)
print(pmem_tx_wa)

#pmem_tx workload b
pmem_tx_wb = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_tx_wb.append(val)
print(pmem_tx_wb)

#pmem_tx workload c
pmem_tx_wc = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_tx_wc.append(val)
print(pmem_tx_wc)

#pmem_tx workload d
pmem_tx_wd = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_tx_wd.append(val)
print(pmem_tx_wd)

#pmem_tx workload e
pmem_tx_we = list()
while(1):
    line = lines[line_it]
    line_it += 1
    if line.startswith("*"): break
    if line.startswith("storeds	workloads/"):
        val = float(line.split('\t')[-1].strip())
        pmem_tx_we.append(val)
print(pmem_tx_we)

######################[pmem-tx data loaded]######################
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, stdev

dram_mean = [mean(dram_wa), mean(dram_wb), mean(dram_wc), mean(dram_wd), mean(dram_we)]
pmem_mean = [mean(pmem_wa), mean(pmem_wb), mean(pmem_wc), mean(pmem_wd), mean(pmem_we)]
pmem_tx_mean = [mean(pmem_tx_wa), mean(pmem_tx_wb), mean(pmem_tx_wc), mean(pmem_tx_wd), mean(pmem_tx_we)]

dram_stdev = [stdev(dram_wa), stdev(dram_wb), stdev(dram_wc), stdev(dram_wd), stdev(dram_we)]
pmem_stdev = [stdev(pmem_wa), stdev(pmem_wb), stdev(pmem_wc), stdev(pmem_wd), stdev(pmem_we)]
pmem_tx_stdev = [stdev(pmem_tx_wa), stdev(pmem_tx_wb), stdev(pmem_tx_wc), stdev(pmem_tx_wd), stdev(pmem_tx_we)]

print(dram_mean)
print(pmem_mean)
print(pmem_tx_mean)

print(dram_stdev)
print(pmem_stdev)
print(pmem_tx_stdev)

# set width of bar
barWidth = 0.25

# Set position of bar on X axis
dram = np.arange(len(dram_mean))
pmem = [x + barWidth for x in dram]
pmem_tx = [x + barWidth for x in pmem]

# Make the plot
plt.bar(dram, dram_mean, color='#034f06', width=barWidth, edgecolor='white', label='dram', yerr=dram_stdev, hatch="//")
plt.bar(pmem, pmem_mean, color='#019999', width=barWidth, edgecolor='white', label='pmem', yerr=pmem_stdev, hatch="--")
plt.bar(pmem_tx, pmem_tx_mean, color='#ff726f', width=barWidth, edgecolor='white', label='pmem_tx', yerr=pmem_tx_stdev, hatch="\\")

# Add xticks on the middle of the group bars
plt.xlabel('skiplist benchmark', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(dram_mean))], ['0/100', '25/75', '50/50', '75/25', '100/0'])

# Add ylim to [0-1200]
plt.ylabel('throughput (KTPS)', fontweight='normal')
plt.ylim([0,1200])

# Create legend & Show graphic
plt.legend()
plt.show()
```
