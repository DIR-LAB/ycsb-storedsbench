```python
# B-tree version compare

import numpy as np
import matplotlib
matplotlib.use("PDF")
import matplotlib.pyplot as plt
import matplotlib.backends as pdf_backend
from statistics import mean, stdev

plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True

pdf = pdf_backend.backend_pdf.PdfPages("msst20_eval_3.pdf")

# set width of bar
barWidth = 0.18

plt.rcParams.update({'font.size': 16})
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

plt.figure(figsize=(10,4))

dram_mean = [(dram_wa[0]), (old_dram_wa[0])]
vmem_mean = [(vmem_wa[0]), (old_vmem_wa[0])]
pmem_mean = [(pmem_wa[0]), (old_pmem_wa[0])]
pmem_tx_mean = [(pmem_tx_wa[0]), (old_pmem_tx_wa[0])]

dram_stdev = [(dram_stdev_wa[0]), (old_dram_stdev_wa[0])]
vmem_stdev = [(vmem_stdev_wa[0]), (old_vmem_stdev_wa[0])]
pmem_stdev = [(pmem_stdev_wa[0]), (old_pmem_stdev_wa[0])]
pmem_tx_stdev = [(pmem_tx_stdev_wa[0]), (old_pmem_tx_stdev_wa[0])]


# Set position of bar on X axis
dram = np.arange(len(dram_mean))
vmem = [x + barWidth for x in dram]
pmem = [x + barWidth for x in vmem]
pmem_tx = [x + barWidth for x in pmem]

# Make the plot
dram_bar = plt.bar(dram, dram_mean, color=(0.1, 0.45, 0.1), width=barWidth, edgecolor='white', label='DRAM', yerr=dram_stdev, capsize=3)
vmem_bar = plt.bar(vmem, vmem_mean, color=(0.5, 0.5, 0.5), width=barWidth, edgecolor='white', label='PMEM-Volatile', yerr=vmem_stdev, capsize=3)
pmem_bar = plt.bar(pmem, pmem_mean, color=(0.13, 0.13, 1), width=barWidth, edgecolor='white', label='PMEM-Persist', yerr=pmem_stdev, capsize=3)
pmem_tx_bar = plt.bar(pmem_tx, pmem_tx_mean, color=(0.9, 0, 0), width=barWidth, edgecolor='white', label='PMEM-Trans', yerr=pmem_tx_stdev, capsize=3)

# Add xticks on the middle of the group bars
# plt.xlabel('B-tree Version Compare', fontweight='bold')
plt.xticks([r + 1.5*barWidth for r in range(len(dram_mean))], ['B-tree Value Referenced', 'B-tree Value Embedded'])
plt.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

# Add ylim to [0-1200]
plt.ylabel('throughput (KTPS)', fontweight='normal')
plt.ylim([0,1200])

# # Add counts above the two bar graphs
for rect in dram_bar + vmem_bar + pmem_bar + pmem_tx_bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, (height + 40), '%d' % int(height), ha='center', va='bottom', rotation=90)

# Create legend
plt.legend(ncol=4, bbox_to_anchor=(0.9, 1.25), fancybox=True, shadow=True, fontsize=12)
# plt.legend(loc='upper center', ncol=6, bbox_to_anchor=(0.0, 1.15), fancybox=True, shadow=True)

# save the plot file
pdf.savefig(bbox_inches = 'tight')

# sub_plt = plt.subplot(1, 1, 1)

# handles,labels = sub_plt.get_legend_handles_labels()

# fig_legend = plt.figure(figsize=(8,.5))
# axi = fig_legend.add_subplot(111)            
# fig_legend.legend(handles, labels, ncol=4, loc='center', scatterpoints = 1)
# axi.xaxis.set_visible(False)
# axi.yaxis.set_visible(False)

# pdf.savefig(bbox_inches = 'tight')
pdf.close()

# Show graphic
# plt.show()
```
