```python
# graph for workload A
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, stdev

array_mean = [mean(array_dram_wa), mean(array_pmem_wa), mean(array_pmem_tx_wa)]
ht_mean = [mean(ht_dram_wa), mean(ht_pmem_wa), mean(ht_pmem_tx_wa)]
btree_mean = [mean(btree_dram_wa), mean(btree_pmem_wa), mean(btree_pmem_tx_wa)]
bp_mean = [mean(bp_dram_wa), mean(bp_pmem_wa), mean(bp_pmem_tx_wa)]
sk_mean = [mean(sk_dram_wa), mean(sk_pmem_wa), mean(sk_pmem_tx_wa)]
rb_mean = [mean(rb_dram_wa), mean(rb_pmem_wa), mean(rb_pmem_tx_wa)]

array_stdev = [stdev(array_dram_wa), stdev(array_pmem_wa), stdev(array_pmem_tx_wa)]
ht_stdev = [stdev(ht_dram_wa), stdev(ht_pmem_wa), stdev(ht_pmem_tx_wa)]
btree_stdev = [stdev(btree_dram_wa), stdev(btree_pmem_wa), stdev(btree_pmem_tx_wa)]
bp_stdev = [stdev(bp_dram_wa), stdev(bp_pmem_wa), stdev(bp_pmem_tx_wa)]
sk_stdev = [stdev(sk_dram_wa), stdev(sk_pmem_wa), stdev(sk_pmem_tx_wa)]
rb_stdev = [stdev(rb_dram_wa), stdev(rb_pmem_wa), stdev(rb_pmem_tx_wa)]

print(array_mean)
print(ht_mean)
print(btree_mean)
print(bp_mean)
print(sk_mean)
print(rb_mean)

print(array_stdev)
print(ht_stdev)
print(btree_stdev)
print(bp_stdev)
print(sk_stdev)
print(rb_stdev)

# set width of bar
barWidth = 0.12

# Set position of bar on X axis
array = np.arange(len(array_mean))
ht = [x + barWidth for x in array]
btree =  [x + barWidth for x in ht]
bp =  [x + barWidth for x in btree]
sk =  [x + barWidth for x in bp]
rb =  [x + barWidth for x in sk]

# Make the plot
array_bar = plt.bar(array, array_mean, color='#3CA606', width=barWidth, edgecolor='white', label='array', yerr=array_stdev, hatch="//")
ht_bar = plt.bar(ht, ht_mean, color='#20CEE9', width=barWidth, edgecolor='white', label='hashtable', yerr=ht_stdev, hatch="--")
btree_bar = plt.bar(btree, btree_mean, color='#F60000', width=barWidth, edgecolor='white', label='b tree', yerr=btree_stdev, hatch="\\")
bp_bar = plt.bar(bp, bp_mean, color='#4E46EF', width=barWidth, edgecolor='white', label='b+ tree', yerr=bp_stdev, hatch="..")
sk_bar = plt.bar(sk, sk_mean, color='#EC60F1', width=barWidth, edgecolor='white', label='skiplist', yerr=sk_stdev, hatch="xx")
rb_bar = plt.bar(rb, rb_mean, color='#7544F5', width=barWidth, edgecolor='white', label='rb tree', yerr=rb_stdev, hatch="++")

# Add xticks on the middle of the group bars
plt.xlabel('workload 0(R)/100(W) benchmark', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(array_mean))], ['dram', 'pmem', 'pmem_tx'])

# Add ylim to [0-1200]
plt.ylabel('throughput (KTPS)', fontweight='normal')
plt.ylim([0,1500])

# Add counts above the two bar graphs
for rect in array_bar + ht_bar + btree_bar + bp_bar + sk_bar + rb_bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, (height + 20), '%d' % int(height), ha='center', va='bottom', rotation=90)

# Create legend
plt.legend(loc='upper center', ncol=6, bbox_to_anchor=(0.5, 1.15), fancybox=True, shadow=True)

# save the plot file
plt.savefig("wla-benchmark.pdf", bbox_inches = 'tight')

# Show graphic
plt.show()
```
