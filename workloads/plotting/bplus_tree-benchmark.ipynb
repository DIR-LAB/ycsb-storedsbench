```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, stdev

dram_mean = [mean(dram_wa), mean(dram_wb), mean(dram_wc), mean(dram_wd), mean(dram_we)]
pmem_mean = [mean(pmem_wa), mean(pmem_wb), mean(pmem_wc), mean(pmem_wd), mean(pmem_we)]
pmem_tx_mean = [mean(pmem_tx_wa), mean(pmem_tx_wb), mean(pmem_tx_wc), mean(pmem_tx_wd), mean(pmem_tx_we)]

dram_stdev = [stdev(dram_wa), stdev(dram_wb), stdev(dram_wc), stdev(dram_wd), stdev(dram_we)]
pmem_stdev = [stdev(pmem_wa), stdev(pmem_wb), stdev(pmem_wc), stdev(pmem_wd), stdev(pmem_we)]
pmem_tx_stdev = [stdev(pmem_tx_wa), stdev(pmem_tx_wb), stdev(pmem_tx_wc), stdev(pmem_tx_wd), stdev(pmem_tx_we)]

print(dram_mean)
print(pmem_mean)
print(pmem_tx_mean)

print(dram_stdev)
print(pmem_stdev)
print(pmem_tx_stdev)

# set width of bar
barWidth = 0.25

# Set position of bar on X axis
dram = np.arange(len(dram_mean))
pmem = [x + barWidth for x in dram]
pmem_tx = [x + barWidth for x in pmem]

# Make the plot
dram_bar = plt.bar(dram, dram_mean, color='#3CA606', width=barWidth, edgecolor='white', label='dram', yerr=dram_stdev, hatch="//")
pmem_bar = plt.bar(pmem, pmem_mean, color='#20CEE9', width=barWidth, edgecolor='white', label='pmem', yerr=pmem_stdev, hatch="--")
pmem_tx_bar = plt.bar(pmem_tx, pmem_tx_mean, color='#F60000', width=barWidth, edgecolor='white', label='pmem_tx', yerr=pmem_tx_stdev, hatch="\\")

# Add xticks on the middle of the group bars
plt.xlabel('b+ tree benchmark', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(dram_mean))], ['0/100', '25/75', '50/50', '75/25', '100/0'])

# Add ylim to [0-1200]
plt.ylabel('throughput (KTPS)', fontweight='normal')
plt.ylim([0,1500])

# Add counts above the two bar graphs
for rect in dram_bar + pmem_bar + pmem_tx_bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, (height + 20), '%d' % int(height), ha='center', va='bottom', rotation=90)

# Create legend
plt.legend(loc='upper center', ncol=6, bbox_to_anchor=(0.5, 1.15), fancybox=True, shadow=True)

# save the plot file
plt.savefig("bplustree-benchmark.pdf", bbox_inches = 'tight')

# Show graphic
plt.show()
```
